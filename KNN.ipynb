```python
import numpy as np
import pandas as pd
import math
```

```python
df=pd.read_csv('ExtraCredit_KNNData.csv')
df
```

```python
d={'No':0,'Yes':1}
df['Continue']=df['Continue?'].apply(lambda x:d[x])
```

```python
xmean=df.x.mean()
ymean=df.y.mean()
print(xmean,ymean)
xstd=df.x.std()
ystd=df.y.std()
print(xstd,ystd)
```

```python
cols = list(df.columns)
cols.remove('ID')
cols.remove('Continue')
cols.remove('Continue?')
cols
```

```python
for col in cols:
    col_zscore = col + '_zscore'
    df[col_zscore] = (df[col] - df[col].mean())/df[col].std()
```

```python
df
```

```python
def knn_mt(x,y,n):
    zero=0
    one=0
    zscore_x=(x-xmean)/xstd
    zscore_y=(y-ymean)/ystd
    for i in df:
        df['Manhattan Distance']= (np.abs(zscore_x-(df['x_zscore'])) + np.abs(zscore_y-(df['y_zscore']))) 
        new_df=df.sort_values(by=['Manhattan Distance'],ascending=False)
        new_df=new_df.head(n)
        #print(new_df)
        for i, row in new_df.iterrows():
            if(row['Continue']==0):
                zero+=1
            else:
                one+=1
    if(zero>one):
        print("By Manhattan taxi-cab distance method:"+"This point is likely a No")
    elif(zero<one):
        print("By Manhattan taxi-cab distance method:"+"This point is likely a Yes")
    else:
        print("By Manhattan taxi-cab distance method:"+"Tie!! Cannot decide") 

def knn_MAX(x,y,n):
    zero=0
    one=0
    zscore_x=(x-xmean)/xstd
    zscore_y=(y-ymean)/ystd
    for i in df:
        df['Maximum Distance']= np.maximum(np.abs(zscore_x-(df['x_zscore'])) , np.abs(zscore_y-(df['y_zscore']))) 
        new_df_max=df.sort_values(by=['Maximum Distance'],ascending=False)
        new_df_max=new_df_max.head(n)
        #print(new_df_max)
        for i, row in new_df_max.iterrows():
            if(row['Continue']==0):
                zero+=1
            else:
                one+=1
    if(zero>one):
        print("By Maximum method:"+"This point is likely a No")
    elif(zero<one):
        print("By Maximum distance method:"+"This point is likely a Yes")
    else:
        print("By Maximum distance method:"+"Tie!! Cannot decide") 
    
def knn(x,y,n):
    knn_mt(x,y,n)
    knn_MAX(x,y,n)
    
# ------------------------------------------ INPUT PARAMETERS BELOW ----------------------------------------------------    


# Enter x, y and no of neighbours  


knn(1,2,4)
```
